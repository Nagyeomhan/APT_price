```python
import pandas as pd
```

```python
df = pd.read_excel('../data/시도별_인구수_데이터.xlsx')
df
```

```python
# 세종특별자치도 '-' 부분을 0으로 변경
df.replace(['-'], 0, inplace=True)
df
```

```python
# 남자, 여자가 포함된 행을 모두 지우기 위해 nan이 포함된 행 모두 drop
df.dropna(inplace=True)
df
```

```python
# 성별 column 없애기
df.drop(['성별'], axis=1, inplace=True)
df
```

```python
# '행정구역별' column '행정구역'으로 이름 바꾸기
df.rename(columns={'행정구역별': '행정구역'}, inplace=True)
df
```

```python
# 인덱스 reset
df.reset_index(drop=True, inplace=True)
df
```

```python
df = pd.melt(df,id_vars=['행정구역'],value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'],
        var_name='년도', value_name='인구수')
df
```

```python
df.to_csv('../data/시도별_인구수_데이터_handling.csv', encoding='utf-8')
```
