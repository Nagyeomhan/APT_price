## 분양가, 인구수, 경제활동인구수, 월평균임금 merge

```python
import pandas as pd
```

```python
price = pd.read_csv('../data/지역별_연도별_평당분양가격.csv', index_col=0, encoding='utf-8')
price.head()
```

```python
workPop = pd.read_csv('../data/경제활동인구_시도_handling.csv', index_col=0, encoding='utf-8')
workPop.head()
```

```python
sidoPop = pd.read_csv('../data/시도별_인구수_데이터_handling.csv', index_col=0, encoding='utf-8')
sidoPop.head()
```

```python
income = pd.read_csv('../data/월평균_임금_handling.csv', index_col=0, encoding='utf-8')
income.head()
```

#### 평당분양가격의 컬럼명 변경

```python
price.rename(columns={'지역명':'행정구역'}, inplace=True)
price
```

#### 행정구역 이름 통일

```python
price.행정구역 = price.행정구역.replace({
    '서울':'서울특별시',
    '부산':'부산광역시',
    '대구':'대구광역시',
    '인천':'인천광역시', 
    '광주':'광주광역시', 
    '대전':'대전광역시', 
    '울산':'울산광역시',       
    '세종':'세종특별자치시', 
    '경기':'경기도', 
    '강원':'강원도', 
    '충북':'충청북도',
    '충남':'충청남도', 
    '전북':'전라북도', 
    '전남':'전라남도', 
    '경북':'경상북도',
    '경남':'경상남도', 
    '제주':'제주특별자치도'
    })
price
```

#### 경제활동인구수의 컬럼명 변경

```python
workPop.rename(columns={'인구수':'경제활동인구'}, inplace=True)
workPop
```

#### 제주도 -> 제주특별자치도

```python
workPop.행정구역 = workPop.행정구역.replace('제주도','제주특별자치도')
workPop
```

#### 년도 -> 연도

```python
workPop.rename(columns={'년도':'연도'}, inplace=True)
workPop
```

```python
sidoPop.rename(columns={'년도':'연도'}, inplace=True)
sidoPop
```

```python
income.rename(columns={'년도':'연도'}, inplace=True)
income
```

#### 분양가-시도별 인구수 merge

```python
df1 = pd.merge(price, sidoPop)
df1
```

#### 경제활동인구수 merge

```python
df2 = pd.merge(df1, workPop)
df2
```

#### 월평균임금 merge

```python
total = pd.merge(df2, income)
total
```

#### csv파일로 저장

```python
total.to_csv('../data/지역별_연도별_비교데이터_합본.csv', encoding='utf-8')
```
