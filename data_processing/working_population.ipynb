## 경제활동인구 데이터 전처리

```python
import pandas as pd
```

```python
df = pd.read_excel('../data/경제활동인구_시도.xlsx')
df
```

### 남자, 여자 index제거

```python
df = df.dropna()
df.head()
```

```python
df.columns
```

### 성별 컬럼 제거

```python
del df['성별']
```

```python
df.head()
```

### 계 index 제거

```python
df = df.drop(index=0)
```

```python
df.head()
```

### reset_index

```python
df = df.reset_index(drop=True)
df.head()
```

### 컬럼명 변경

```python
df.rename(columns={'시도별':'행정구역'}, inplace=True)
df
```

### Melt

```python
df = pd.melt(df, id_vars=['행정구역'], value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'], var_name='년도', value_name='인구수')
df
```

### nan값 처리

```python
df.인구수 = df.인구수.replace(['-'],0)
df.info()
```

### 천명 -> 명 단위변경

```python
df.인구수 = df.인구수*1000
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_csv('../data/경제활동인구_시도_handling.csv',encoding='utf-8')
```
